Data Preprocessing and Building Models

1. Data Preprocessing

```python
# counts of unique rows in the 'diagnosis' column
df['diagnosis'].value_counts()
```

```python
# mapping categorical values to numerical values
df['diagnosis']=df['diagnosis'].map({'B':0,'M':1})
```

```python
df['diagnosis'].value_counts()
```

2. Splitting the data into train and test

```python
from sklearn.model_selection import train_test_split

# splitting data
X_train, X_test, y_train, y_test = train_test_split(
                df.drop('diagnosis', axis=1),
                df['diagnosis'],
                test_size=0.2,
                random_state=42)

print("Shape of training set:", X_train.shape)
print("Shape of test set:", X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.fit_transform(X_test)
```

StandardScaler menstandardisasi fitur dengan mengurangkan mean dan kemudian menskalakan ke varians unit. (Varians unit berarti membagi semua nilai dengan standar deviasi.)

SVM Classification Model

```python
from sklearn.svm import SVC

svc_model = SVC(kernel="rbf")
svc_model.fit(X_train, y_train)
predictionSVM = svc_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print("Confusion Matrix: \n", confusion_matrix(y_test, predictionSVM))
print("\n")
print(classification_report(y_test, predictionSVM))
```

```python
from sklearn.metrics import accuracy_score
svm_acc = accuracy_score(y_test, predictionSVM)
print("Accuracy of SVM model is: ", svm_acc)
```
